## Sucesión Aritmética: Fórmula


**Fórmula:**  

aₙ = a₁ + (n−1)d

**Ejemplo:**  
Si a₁ = 10 y d = 4, entonces:
a₅ = 10 + (5−1)*4 = 26.


## Sucesión Geométrica: Fórmula


**Fórmula:**  

aₙ = a₁ · rⁿ⁻¹

**Ejemplo:**  
Si a₁ = 3 y r = 2, entonces:
a₄ = 3·2³ = 24.


## Multiplicación de Matrices


**Fórmula:**  
C = A·B

**Ejemplo:**
A=[[1,2],[3,4]], B=[[2,0],[1,2]]
C=[[1*2+2*1, 1*0+2*2],[3*2+4*1,3*0+4*2]]=[[4,4],[10,8]].

## Matriz Inversa


**Fórmula:**
