En época invernal, es común que algunas ciudades experimenten lluvias intensas, lo que puede poner en riesgo su infraestructura. Este año, ciudades como Valparaíso, Santiago y Concepción han registrado precipitaciones que requieren atención especial.

Durante un día de lluvia podemos representar la cantidad de agua acumulada, en milímetros (mm), en cada ciudad con respecto al tiempo transcurrido en horas (t) desde las 6 de la mañana, a través de las siguientes funciones cuadráticas:

Valparaíso: 

Y(t) = 0,94t^2 + 12,5t + 38

Santiago: 

S(t) = 0,96t^2 + 16,8t + 31

Concepción: 

C(t) = 1,1t^2 + 19,4t + 26

Se debe considerar que las redes de alcantarillado de las ciudades no pueden acumular más de 100 mm de agua. Si se supera este umbral, será necesario enviar equipos de emergencia para evitar una inundación.

1. ¿Cuál es el valor de t a las 07:39 de la mañana? ¿Cuál sería la cantidad de agua acumulada en Concepción a las 07:39 de la mañana?

```python
#Concepción: 
def c(t):
    return 1.1*t**2 + 19.4*t + 26

valor = 99/60
print(f"El valor de t a las 07:39 am es de {valor:.2f}")
print(f"La cantidad de agua acumulada en Concepción a las 07:39 de la mañana es de {c(valor):.0f} mm")

```

2. ¿Cuál será la máxima cantidad de agua acumulada considerando que se pronostican 14 horas de lluvia?

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import fsolve

def Y(t):
    return 0.94*t**2 + 12.5*t + 38 #Valparaíso

def S(t):
    return 0.96*t**2 + 16.8*t + 31 #Santiago

def C(t):
    return 1.1*t**2 + 19.4*t + 26 #Concepción 

print(f"El agua acumulada en Valparaiso en 14 horas de lluvia es de {Y(14):.0f} mm")
print(f"El agua acumulada en Santiago en 14 horas de lluvia es de {S(14):.0f} mm")
print(f"El agua acumulada en Concepción en 14 horas de lluvia es de {C(14):.0f} mm")

# GRAFICAR FUNCIONES: 
t = np.arange(0, 14, 0.1) # (Inicio, fin, paso)
plt.plot(t, Y(t), label="Valparaiso Tiempo transcurrido en horas")  #dibuja una curva x = t, y = f(t)
plt.plot(t, S(t), label="Santiago Tiempo transcurrido en horas")  #dibuja una curva x = t, y = f(t) 
plt.plot(t, C(t), label="Concepción Tiempo transcurrido en horas")  #dibuja una curva x = t, y = f(t) 

# Ajustamos los ejes: [xmin, xmax, ymin, ymax]
# plt.axis([0, 1600, 0, 30000])

plt.legend() #muestra la leyenda con los nombres que pusimos en label
plt.xlabel("Tiempo transcurrido en horas (t)") #nombre del eje x
plt.ylabel("Cantidad de agua almacenada en mm f(t)") #nombre del eje y
plt.title("Relacion cantidad de agua almacenada en tiempo transcurrida") #título del gráfico
plt.grid(True) #activa la rejilla (cuadrícula) para ver mejor los valores
plt.show() # muestra el grafico. a veces no lo muestra sin este codigo
```

3. ¿Cuál será el primer momento en que se necesitará el apoyo de emergencia?

```python
from scipy.optimize import fsolve
# DESPEJAR FUNCIONES A 0. f(t) toma el valor de 100
# EJ: 100 = 0.94*t**2 + 12.5*t + 38

def V(t):
    return 0.94*t**2 + 12.5*t + 38 - 100 #Valparaíso

def S(t):
    return 0.96*t**2 + 16.8*t + 31 - 100 #Santiago

def C(t):
    return 1.1*t**2 + 19.4*t + 26 - 100 #Concepción 

solucion_v = fsolve(V, 100) # SOLUCION DE VALPARAISO
solucion_s = fsolve(S, 100) # SOLUCION DE SANTIAGO
solucion_c = fsolve(C, 100) # SOLUCION DE CONCEPCION

print(f"La cantidad de horas donde comenzará a pedir apoyo de emergencia Valparaiso es de: {solucion_v[0]:.1f} horas")
print(f"La cantidad de horas donde comenzará a pedir apoyo de emergencia Santiago es de: {solucion_s[0]:.1f} horas")
print(f"La cantidad de horas donde comenzará a pedir apoyo de emergencia Concepcion es de: {solucion_c[0]:.1f} horas")
print(f"\nEl primero en pedir apoyo de emergencia es Concepcion con {solucion_c[0]:.1f} horas ")
```

4. Entre las 3 y 5 de la tarde, ¿en qué ciudad cayó más agua?

```python
def V(t):
    return 0.94*t**2 + 12.5*t + 38 #Valparaíso

def S(t):
    return 0.96*t**2 + 16.8*t + 31 #Santiago

def C(t):
    return 1.1*t**2 + 19.4*t + 26 #Concepción 

# De 6 a 15 hay 9 horas
# De 6 a 17 hay 11 horas
# Hay que evaluar [9, 11] en cada uno, y la diferencia que hay en ese rango horario

print(f"La diferencia de agua en Valparaiso es de: {V(11) - V(9):.1f} mm")
print(f"La diferencia de agua en Santiago es de: {S(11) - S(9):.1f} mm")
print(f"La diferencia de agua en Concepción es de: {C(11) - C(9):.1f} mm")
print(f"\nDonde más llovió fue en Concepción con una diferencia de {C(11) - C(9):.1f} mm entre las 15 y 17 horas")
```
