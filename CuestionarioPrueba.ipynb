Sucesi√≥n aritm√©tica: $$a_{n} = a_{1} + (n - 1)\cdot d$$

### Sea una sucesi√≥n cuyos tres primeros t√©rminos son 200, 224 y 248. Si la sucesi√≥n mantiene el mismo patr√≥n de crecimiento sucesivamente.

##### Pregunta 1: El valor del t√©rmino que ocupa el lugar 36 en esta sucesi√≥n es 

```python
def A(n):
    return a1 + (n-1)*d

a1 = 200 # Primer termino de la sucesi√≥n
d = 224 - 200

print(A(36))
```

```python
#for
lista = []

for n in range(1, 37):
    lista.append(A(n)) 

print(sum(lista))    
```
