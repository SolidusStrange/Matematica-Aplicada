Sucesión aritmética: $$a_{n} = a_{1} + (n - 1)\cdot d$$

```python
# Sea una sucesión cuyos tres primeros términos son 200, 224 y 248. Si la sucesión mantiene el mismo patrón de crecimiento sucesivamente.
```

```python
# Pregunta 1
# El valor del término que ocupa el lugar 36 en esta sucesión es 
def A(n):
    return a1 + (n-1)*d

a1 = 200 # Primer termino de la sucesión
d = 224 - 200

print(A(36))
```

```python
# Pregunta 2
# ¿Cuál es la suma de los primeros 36 términos de esta sucesión? 
lista = []

for n in range(1, 37):
    lista.append(A(n)) 

print(sum(lista))    
```

```python
# 5) Si el rendimiento de los servidores crece a razón constante, ¿cuál es el rendimiento de un servidor el noveno mes? 
def A(n):
    return a1 * r**(n-1)


a2 = 1066
a5 = 8528

r = (a5/a2) ** (1/3)
print(r)

a1 = (a2 / r) # Determinamos a1

print(A(9))
```

```python
# 6) Si el rendimiento de los servidores crece a razón constante, ¿en qué mes el rendimiento de un servidor es de 34.112 TPS? 

servidores = []
valor_buscado = 34112

for n in range(1, 100):
    servidores.append(A(n))
    
if valor_buscado in servidores:
    print(servidores.index(valor_buscado)+1)
else:
    print("El valor buscado no se encuentra en la lista")
```
