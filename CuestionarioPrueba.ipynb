Sucesión aritmética: $$a_{n} = a_{1} + (n - 1)\cdot d$$

### Sea una sucesión cuyos tres primeros términos son 200, 224 y 248. Si la sucesión mantiene el mismo patrón de crecimiento sucesivamente.

##### Pregunta 1: El valor del término que ocupa el lugar 36 en esta sucesión es 

```python
def A(n):
    return a1 + (n-1)*d

a1 = 200 # Primer termino de la sucesión
d = 224 - 200

print(A(36))
```

##### Pregunta 2: ¿Cuál es la suma de los primeros 36 términos de esta sucesión? 

```python
#for
lista = []

for n in range(1, 37):
    lista.append(A(n)) 

print(sum(lista))    
```
