# Problema 1

Una red transmite datos a $100$ megabits por segundo. Crea una función utilizando *Python* que permita calcular la cantidad de datos transmitidos en una cierta cantidad de tiempo (en segundos). Luego utiliza la función y calcula cuántos datos se transmiten en:

* $45$ segundos
* $1,5$ minutos
* $1$ hora



**Respuesta:** Primero hay que crear la función determinando cuál es la variable dependiente e independiente: 
- Variable independiente ($t$): tiempo en segundos
- Variable dependiente $m(t)$: 100mbs transmitidos por segundo

Luego se realizan los cálculos teniendo en cuenta que 100mbs son por segundo → 100 * $t$
- Por lo que la función sería: $m(t)$: 100t

```python
def m(t):
  return 100*t

# 45 segundos
# 1,5 minutos = 90 segundos
# 1 hora = 60 min = 3600 segundos

print(f"A los 45 segundos se transmiten {m(45):.0f} megabits")
print(f"A los 1.5 minutos segundos se transmiten {m(1.5*60):.0f} megabits")
print(f"A la hora se transmiten {m(60*60):.0f} megabits")
```

# Problema 5

Un turista ha llegado a Santiago y desea conocer algunos lugares de la ciudad. Ha decidido visitar el Palacio de la Moneda y desde ahí trasladarse al centro comercial "Costanera Center", utilizando algún medio de transporte que ofrece la ciudad.

Si se traslada en metro deberá abordar en estación La Moneda y bajar en la estación Tobalaba ($9$ estaciones). La función $f(t)$ permite calcular la distancia recorrida utilizando el metro (en kilómetros) transcurridos $t$ minutos.

$$f(t) = 0,4t$$

Si se traslada en bus, el turista podrá observar la ciudad y otros atractivos en su viaje. La función $g(t)$ permite calcular la distancia recorrida en bus (en kilómetros) transcurridos $t$ minutos.

$$g(t) = 0,3t$$

Según la información anterior:

1. Grafique ambas funciones, indicando el nombre de cada eje junto con su unidad de medida. Para realizar el gráfico utilice la biblioteca *Matplotlib*.
2.  Si se sabe que el metro se demora $1,2$ minutos en llegar desde una estación a otra y espera $30$ segundos en cada estación, indique el dominio contextualizado para $f(t)$.
3. Mediante análisis gráfico, indique cuál medio de transporte es más conveniente en términos de tiempo, para el turista. Justifique.
4.  Si se sabe que desde estación La Moneda hasta Tobalaba son aproximadamente $6$ kilómetros ¿cuántos tiempo tardará el turista en llegar a su destino con cada una de las opciones?

**#PREGUNTA 5.1:** Grafique ambas funciones, indicando el nombre de cada eje junto con su unidad de medida. Para realizar el gráfico utilice la biblioteca *Matplotlib*

```python
#PREGUNTA NÚMERO 5.1
import matplotlib.pyplot as plt #es el módulo de gráficas en Python. Se abrevia como plt por convención.
import numpy as np #es una librería para trabajar con vectores, matrices y rangos numéricos. Se abrevia np

# Definimos las funciones
def f(t):
    return 0.4 * t  # metro

def g(t):
    return 0.3 * t  # bus

# Rango de tiempo (0 a 15 minutos)
t = np.arange(0, 15, 0.1)  #np.arange(inicio, fin, paso)

# Graficar ambas funciones
plt.plot(t, f(t), label="Metro (f(t) = 0.4t)") #dibuja una curva x = t, y = f(t)
plt.plot(t, g(t), label="Bus (g(t) = 0.3t)")
plt.axhline(y=6, color='r', linestyle='--', label="Distancia 6 km")
plt.legend() #muestra la leyenda con los nombres que pusimos en label
plt.xlabel("Tiempo en minutos") #nombre del eje x
plt.ylabel("Distancia recorrida en KM") #nombre del eje y
plt.title("Recorrido del Metro") #título del gráfico
plt.grid(True) #activa la rejilla (cuadrícula) para ver mejor los valores
plt.show() #muestra el grafico. a veces no lo muestra sin este codigo
```

**#PREGUNTA 5.2** Si se sabe que el metro se demora $1,2$ minutos en llegar desde una estación a otra y espera $30$ segundos en cada estación, indique el dominio contextualizado para $f(t)$

**R**: Dominio contextualizado para f(𝑡)
Entre La Moneda y Tobalaba son 9 tramos de viaje (de una estación a la siguiente).
*Tiempo en movimiento:* 9×1,2=10,8 min

*Esperas en estaciones intermedias:* hay 8 paradas intermedias → 8×0,5=4 min

*Tiempo total del recorrido en metro:* 10,8+4 → 14,8 min

*Dom f(t): [0, 14.8] minutos*

**#PREGUNTA 5.3**
¿Qué medio conviene en tiempo?

**R**: La recta del metro tiene mayor pendiente (0,4 km/min) que la del bus (0,3 km/min), por lo que para una misma distancia el metro alcanza antes esa distancia. *Conviene el metro en términos de tiempo*.

**#PREGUNTA 5.4:** Resolver ecuación

**R**: Tiempo para recorrer *6KM*
Usando las funciones:
    - Metro 0.4t = 6 → 6/0.4 = 15min
    - Bus 0.3t = 6 → 6/0.3 = 20min

- Para ir desde La Moneda al Costanera Center (~6 km), el metro demora ~15 min, el bus ~20 min.

```python
#PREGUNTA 5.4: PRACTICA DE FSOLVE
import numpy as np
from scipy.optimize import fsolve

def f(t):
    return 0.4 * t - 6 # metro

def g(t):
    return 0.3 * t - 6  # bus

t= np.linspace(0, 25, 1) #RANGO SIMILAR A ARANGE PERO VALORES QUE PODRIA TOMAR t. NO CONSIDERA ULTIMO ELEMENTO
solucion_f = fsolve(f, t) #primero la funcion, despues la variable independiente
solucion_g = fsolve(g, t) #primero la funcion, despues la variable independiente
print(f"Se demora {solucion_f[0]:.0f} minutos en recorrer 6 kms en Metro") #0 porque queremos solo una solución
print(f"Se demora {solucion_g[0]:.0f} minutos en recorrer 6 kms en Bus") #0 porque queremos solo una solución
```

# Problema 7

El número de usuarios de una red social se puede modelar mediante la función:

$$U(t) = \frac{1.000}{1+9e^{-0.5t}}$$

donde $t$ corresponde al tiempo transcurrido en meses.

1.  Defina variable dependiente e independiente, indicando unidad de medida.
2.  Determine la cantidad de usuarios transcurridos $12$ meses.
3.  Grafique la función, utilizando la librería *Matplotlib*, para los primeros dos años de funcionamiento.
4.  ¿Cuánto tiempo debe pasar para que la red social llegue a $800$ usuarios?

**Pregunta 7.1: Defina variable dependiente e independiente, indicando unidad de medida.**

- Variable independiente ($t$): tiempo transcurrido en meses
- Variable dependiente $U$($t$): numero de usuarios de una red social en unidades

**Pregunta 7.2: Determine la cantidad de usuarios transcurridos 12 meses** 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def U(t):
  return 1000/(1 + 9*np.exp(-0.5*t))

print(f"A los 12 meses tendrá {U(12):.0f} usuarios la red social")
```

**Pregunta 7.3: Grafique la función, utilizando la librería *Matplotlib*, para los primeros dos años de funcionamiento.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def U(t):
  return 1000/(1 + 9*np.exp(-0.5*t))

t = np.arange(0, 24, 0.1) #Dom(t):[0,24] primeros dos años

plt.plot(t, U(t),label= "Tiempo")
plt.legend()
plt.xlabel("Tiempo transcurrido en meses")
plt.ylabel("Numero de usuarios en unidades")
plt.title("Numero de usuarios de una red social")
plt.grid(True)
```

**Pregunta 7.4: ¿Cuánto tiempo debe pasar para que la red social llegue a $800$ usuarios?**

```python
#PREGUNTA 4
import numpy as np
from scipy.optimize import fsolve

def U(t):
  return 1000/(1 + 9*np.exp(-0.5*t))-800 # despejar la ecuación

t= np.linspace(0, 25, 1) # RANGO SIMILAR A ARANGE PERO VALORES QUE PODRIA TOMAR t. NO CONSIDERA ULTIMO ELEMENTO
solucion = fsolve(U, t) # Primero la funcion, despues la variable independiente
print(f"A los {solucion[0]:.1f} meses aproximadamente serán 800 usuarios") #0 porque queremos solo una solución
```
