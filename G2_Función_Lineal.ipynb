# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


Respuesta 1:
- La variable independiente: Es la cantidad de elementos de entrada en unidades
- La variable dependiente: Es el tiempo de ejecución en ms

```python
# Respuesta 1:
print("Respuesta1:\nLa variable independiente: Es la cantidad de elementos de entrada en unidades\nLa variable dependiente: es el tiempo de ejecución en ms ")

# Respuesta 2:
import numpy as np
from scipy.optimize import fsolve 

# Datos
x = np.array([100, 200, 500, 1000, 2000])  # elementos
y = np.array([2, 4, 10, 20, 40])           # tiempo en ms 

# Ajuste lineal con polyfit
a, b = np.polyfit(x , y, 1)
print(f"Respuesta 2: La funcion es f(x) = {a:.2f}x + {b:.2f}")

# Respuesta 3
# Tiempo para 1500 elementos
t_1500 = a * 1500 + b
print(f"Respuesta 3: Tiempo para 1500 elementos: {t_1500:.2f} ms")

# Respuesta 4
# Resolver con fsolve el tamaño de entrada para T(n) = 50
def ecuacion(n): # Definir la ecuación
    return a*n + b - 50 # Queremos que T(n) = 50

# Valor inicial para que fsolve arranque
n0 = np.linspace(100, 3000, 1)

# Resolver
solucion = fsolve(ecuacion, n0)
print(f"Respuesta 4: Tamaño de entrada para que T=50 ms: {solucion[0]:.0f} elementos")
```

1. Defina variable dependiente e independiente, indicando unidad de medida.

3. Interprete la pendiente en el contexto del problema.
4. ¿Cuánto tiempo toma una transferencia de $73,2$ GB?

```python
# Respuesta 1
print("Respuesta 1: La variable independiente es datos transferidos en gb\nLa variable dependiente: tiempo de transferencia en minutos ")

# Respuesta 2
import numpy as np
from scipy.optimize import fsolve

# Datos
x = np.array([5, 10, 25, 50, 100])  # datos transferidos
y = np.array([10, 20, 50, 100, 200]) # tiempo de transferencia

# Ajuste lineal con polyfit
a, b = np.polyfit(x , y, 1)
print(f"Respuesta 2: La funcion es T(x) = {a:.2f}x + {b:.2f}")

# Respuesta 3
print("Respuesta 3: La pendiente indica que por cada dato transferido el tiempo de transferencia aumenta en 2")

# Respuesta 4
def T(x):
    return 2 * x + 0

print(f"Respuesta 4: El tiempo que toma hacer una transferencia de 73,2GB es de {T(73.2):.2f} minutos")

# Respuesta 5
def ecuacion(x): # Definimos la ecuacion
    return 2 * x - 123.5 # Queremos que T(x) = 123.5

# Valor inicial para que fsolve arranque
x0 = np.linspace(100, 200, 1)

# Resolver
solucion = fsolve(ecuacion, x0)
print(f"Respuesta 5: La cantidad de datos que se pueden transferir en 123,5 minutos es de {solucion[0]:.2f} GB")
 
# Respuesta 6
import matplotlib.pyplot as plt

def T(x):
  return 2*x + 0

x = np.arange(0, 100, 0.1)
plt.plot(x, T(x), label = "Tiempo en minutos de transferencia de datos en gb")
plt.grid(True)
plt.xlabel("Datos transferidos (GB)")
plt.ylabel("Tiempo de transferencia (Minutos)")
plt.legend()
plt.show()
print("Respuesta 6: Que es una pendiente creciente que indica que a medida que los datos transferidos aumentan, también lo hará el tiempo de la transferencia")

```

$$g(t) = 0,7t^{2}$$

mientras que la distancia recorrida por el atleta $2$ está dada por la función:

$$f(t) = 10t $$

donde $t$ corresponde al tiempo transcurrido (en segundos) desde el inicio del trayecto.

1. ¿Cuál de los dos atletas siempre mantuvo una velocidad constante? Justifique. *Recomendación: grafica las funciones utilizando Matlplotlib.*

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def g(t):
  return 0.7 * t**2

def f(t):
  return 10 * t

t = np.arange(0, 10, 0.1)
plt.plot(t, g(t), label = "Atleta 1")
plt.plot(t, f(t), label = "Atleta 2")
plt.grid(True)
plt.xlabel("Tiempo transcurrido en segundos")
plt.ylabel("Distancia recorrida en metros")
plt.legend()
plt.show()
print("El atleta 2 es quien mantiene una velocidad constante")

```
