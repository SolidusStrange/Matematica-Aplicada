# Problema 4

El uso de memoria $M(x)$ (en GB) de un servidor en función del número de usuarios activos $x$ está dado por la función lineal:

 $$M(x) = 0.5x + 2$$

1. Determina la variable dependiente e independiente, indicando unidad de medida.
2. Identifica e interpreta la pendiente de la función.
3. ¿Cuánta memoria del servidor está en uso cuando no hay usuarios activos?
4. ¿Cuánta memoria del servidor está en uso cuando hay $637$ usuarios activos?
5. Utiliza *Python* para determinar cuántos usuarios activos provocan un uso de memoria de $32$ GB.
6. ¿Es posible un uso de memoria de $59,8$ GB?

1. **Determina la variable dependiente e independiente, indicando unidad de medida**
- Variable independiente: Número de usuarios activos unidades (x)
- Variable dependiente: Uso de memoria en GB M(x)

2. **Identifica e interpreta la pendiente de la función.**
- m = 0.5 (Es creciente, y como es positiva el uso de la memoria por cada usuario activo aumenta en 0.5 GB)

3. **¿Cuánta memoria del servidor está en uso cuando no hay usuarios activos?**
- M(0) = 0.5 * 0 + 2 → 2

```python
#3 ¿Cuánta memoria del servidor está en uso cuando no hay usuarios activos?
def m(x):
    return 0.5 * x + 2

print(f"El uso de memoria cuando no hay usuarios es de {m(0)} GB")

#4 ¿Cuánta memoria del servidor está en uso cuando hay $637$ usuarios activos?

print(f"El uso de memoria cuando hay 637 es de {m(637)} GB")
```

```python
#5 Utiliza *Python* para determinar cuántos usuarios activos provocan un uso de memoria de $32$ GB
from scipy.optimize import fsolve
import numpy as np

def independiente(x):
    return 0.5 * x + 2 - 32

x = np.linspace(0, 100, 1)
solucion = fsolve(independiente, x)
print(f"La cantidad de usuario es de {solucion[0]:.0f} para que el uso de memoria sea de 32GB")
```

```python
# 6. ¿Es posible un uso de memoria de $59,8$ GB?
from scipy.optimize import fsolve
import numpy as np

def independiente(x):
    return 0.5 * x + 2 - 59.8

x = np.linspace(0, 100, 1)
solucion = fsolve(independiente, x)
print(f"La cantidad de usuario es de {solucion[0]} para que el uso de memoria sea de 59.8GB")

# Al tener 115.6 usuarios no se puede tener ese uso de memoria
```

# Problema 5

El costo de operación mensual $C(x)$ (en miles de dólares) de un data center en función del número de servidores $x$ está dado por la función lineal:

 $$C(x) = 1,5x + 10$$

1. Determine la variable dependiente e independiente, indicando unidad de medida.
2. ¿Qué representa la pendiente en el contexto del problema?
3. ¿Qué representa el intercepto en el contexto del problema?
4. Calcula el costo de operación cuando hay $47$ servidores.
5. Determina cuántos servidores hay cuando el costo de operación es de $92.500$ dólares. Utiliza *Python* para resolver.


**1. Determine la variable dependiente e independiente, indicando unidad de medida.**
- Variable dependiente: costo de operación mensual en miles de dolares C(x)
- Variable independiente: número de servidores en unidades (x)

**2. ¿Qué representa la pendiente en el contexto del problema?**
- La pendiente es creciente porque es positiva y aumenta 1.5 por cada unidad de número de servidores

**3. ¿Qué representa el intercepto en el contexto del problema?**
- El intercepto indica que a las 10 unidades choca con el eje y, por lo tanto comienza desde 10 unidades de servidor

```python
# 4. Calcula el costo de operación cuando hay $47$ servidores.

def C(x):
    return 1.5 * x + 10

print(f"El costo de operación cuando hay 47 servidores es de {C(47)} miles de dolares")
```

```python
# 5. Determina cuántos servidores hay cuando el costo de operación es de $92.500$ dólares. Utiliza *Python* para resolver.
import numpy as np
from scipy.optimize import fsolve

def variable(x):
    return 1.5 * x + 10 - 92.5

x = np.linspace(0, 1000, 1)
solucion = fsolve(variable, x)
print(f"La cantidad de servidores cuando el costo de operacion es de 92.500 dolares es de {solucion[0]:.0f} unidades")
```

**1. Defina variable dependiente e independiente del problema, indicando unidad de medida.**
- Variable dependiente: visitas diarias en un sitio web
- Variable independiente: número de campañas publicitarias

**2. Determine el dominio contexualizado de la función**
- Dom = [0, 20]

**3. Interprete la pendiente de la función en el contexto del problema.**
- La pendiente es creciente porque es positiva, y aumenta en 300 por cada campaña publicitaria

# Problema 10

Dos atletas profesionales corren por un camino recto. Durante los primeros $10$ segundos del trayecto, la distancia recorrida (en metros) por el atleta $1$ está dada por la función:

$$g(t) = 0,7t^{2}$$

mientras que la distancia recorrida por el atleta $2$ está dada por la función:

$$f(t) = 10t $$

donde $t$ corresponde al tiempo transcurrido (en segundos) desde el inicio del trayecto.

<br>

1. ¿Cuál de los dos atletas siempre mantuvo una velocidad constante? Justifique. *Recomendación: grafica las funciones utilizando Matlplotlib.*
2. Si la meta se encontraba a $100$ metros de la partida ¿cuál atleta llega primero?
3. ¿A qué distancia se encontraban uno del otro cuando transcurren $8$ segundos? ¿Y cuando transcurren $10$ segundos?
4. Entre el segundo $8$ y el $10$, ¿cuál atleta iba más rápido? Justifique apropiadamente.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def g(t):
  return 0.7 * t**2

def f(t):
  return 10 * t

t = np.arange(0, 10, 0.1)
plt.plot(t, g(t), label = "Atleta 1")
plt.plot(t, f(t), label = "Atleta 2")
plt.grid(True)
plt.xlabel("Tiempo transcurrido en segundos")
plt.ylabel("Distancia recorrida en metros")
plt.legend()
plt.show()
print("El atleta 2 es quien mantiene una velocidad constante")

```

```python
def g(t):
  return 0.7 * t**2 - 100

def f(t):
  return 10 * t - 100

t = np.linspace(0, 1000, 1)
solucion1 = fsolve(g, t)
solucion2 = fsolve(f, t)

if solucion1>solucion2:
  print(f"El atleta 1 llegó primero")
else:
  print("El atleta 2 llegó primero")




```
