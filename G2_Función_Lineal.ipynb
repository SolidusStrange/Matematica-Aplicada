# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


Respuesta 1:
- La variable independiente: Es la cantidad de elementos de entrada en unidades
- La variable dependiente: Es el tiempo de ejecución en ms

```python
# Respuesta 1:
print("Respuesta1:\nLa variable independiente: Es la cantidad de elementos de entrada en unidades\nLa variable dependiente: es el tiempo de ejecución en ms ")

# Respuesta 2:
import numpy as np
from scipy.optimize import fsolve 

# Datos
x = np.array([100, 200, 500, 1000, 2000])  # elementos
y = np.array([2, 4, 10, 20, 40])           # tiempo en ms 

# Ajuste lineal con polyfit
a, b = np.polyfit(x , y, 1)
print(f"Respuesta 2: La funcion es f(x) = {a:.2f}x + {b:.2f}")

# Respuesta 3
# Tiempo para 1500 elementos
t_1500 = a * 1500 + b
print(f"Respuesta 3: Tiempo para 1500 elementos: {t_1500:.2f} ms")

# Respuesta 4
# Resolver con fsolve el tamaño de entrada para T(n) = 50
def ecuacion(n): # Definir la ecuación
    return a*n + b - 50 # Queremos que T(n) = 50

# Valor inicial para que fsolve arranque
n0 = np.linspace(100, 3000, 1)

# Resolver
solucion = fsolve(ecuacion, n0)
print(f"Respuesta 4: Tamaño de entrada para que T=50 ms: {solucion[0]:.0f} elementos")
```

# Problema 7

La depreciación es el proceso en que los activos físicos, como la maquinaria, vehículos, equipos, etc., pierden valor con el tiempo debido a factores como el desgaste, el uso y la obsolescencia. El valor estimado que tendrá el activo al final de su vida útil se conoce como *valor residual*.

En este contexto, sabemos que el valor residual $R(t)$ (en miles de dólares) de un equipo tecnológico está modelado por la función lineal:

$$R(t) = -1,2t + 10$$

donde $t$ corresponde a la antigüedad del equipo (en años).

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. ¿Qué representa la pendiente de la función en el contexto del problema?
3. ¿Cuál es el valor residual inicial del equipo?
4. Determine el dominio contextualizado de la función, si sabe que el valor del equipo al final de su vida útil es de $400$ dólares.
5. Determina, utilizando la función *def* de *Python*, el valor residual del equipo con $3$ años y seis meses de antigüedad.

7. Grafica la función utilizando la biblioteca *Matplotlib*. Interpreta lo observado.

```python
import numpy as np
from scipy.optimize import fsolve

# Respuesta 1
print("Respuesta 1: La variable dependiente es el valor residual de un equipo tecnológico en miles de dolares \
y la variable independiente es la antiguedad del equipo en años")

# Respuesta 2
print("Respuesta 2: La pendiente representa que por cada año de antiguedad del equipo este se deprecia en 1200 dolares o el valor residual disminuye")

# Respuesta 3
print("Respuesta 3: El valor residual inicial del equipo depende del intercepto, en este caso es de 10000 dolares")

# Respuesta 4
def ecuacion(t): # Definiminos la ecuacion
    return -1.2*t + 10 - 0.4 # Queremos que R(t) = (400/1000)

t0 = np.linspace(0, 20, 1) # Valor inicial para que fsolve arranque

solucion = fsolve(ecuacion, t0) # Resolver

print(f"Respuesta 4: El dominio contextualizado para que el valor del equipo al final de su vida util sea de 400 dolares es de [0, {solucion[0]}].\
 Por lo que después de 8 años el equipo alcanza el valor de 400 dolares")

# Respuesta 5
def R(t):
    return -1.2*t + 10

print(f"Respuesta 5: El valor residual del equipo con 3 y seis meses de antiguedad es de {R(3.5)*1000} dolares")

# Respuesta 6
def ecuacion(t): # Definimos la ecuación
    return -1.2*t + 10 - 4 # Queremos R(t) = 4

t0 = np.linspace(0, 20, 1) # Valor inicial para que fsolve arranque

solucion = fsolve(ecuacion, t0) # Resolver

print(f"Respuesta 6: La antiguedad del equipo cuando su valor residual es de 4000 dolares es de {solucion[0]} años")

# Respuesta 7
print("Respuesta 7: El valor residual inicia en 10000 dolares, baja linealmente de 1200 dolares por año y llega a $400 dolares a los 8 años.\nDespués de 8 años el modelo da valores negativos, lo cual no tiene sentido físico por eso el dominio se restringe a [0 , 8]")
import matplotlib.pyplot as plt

def R(t):
  return -1.2*t + 10

t = np.arange(0, 20, 0.1)
plt.plot(t, R(t), label = "Valor residual del equipo en miles de dolares")
plt.grid(True)
plt.xlabel("Antiguedad del equipo en años")
plt.ylabel("Valor residual del equipo en miles de dolares")
plt.legend()
plt.show()




    




```

# Problema 11

Un *Data Center* ha cotizado el costo de desarrollar un proyecto con dos empresas. El costo (en millones de pesos) entregado por la empresa $1$ está modelado por la función:

$$C_{1}(t) = 12t + 50$$

y costo (en millones de pesos) entregado por la empresa $2$ está modelado por la función:

$$C_{2}(t) = 8t + 100$$

donde $t$ corresponde al tiempo que demoren en realizar el proyecto (en semanas).

<br>

1. Utilizando la biblioteca *Matplotlib* grafique ambas funciones, indicando el nombre de los ejes coordenados junto con su unidad de medida.
2. Si con la empresa $1$ pagaron $155$ millones de pesos ¿cuántas semanas duró el trabajo?
3. Analizando el gráfico de la pregunta $1$, determine: ¿cuándo conviene desarrollar el proyecto con la empresa $1$? ¿Y la empresa $2$?

```python
# Respuesta 1
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

print("Respuesta 1:")

def C1(t):
  return 12*t + 50

def C2(t):
  return 8*t + 100

t = np.arange(0, 100, 0.1)

plt.plot(t, C1(t), label = "Empresa 1")
plt.plot(t, C2(t), label = "Empresa 2")
plt.grid(True)
plt.xlabel("Tiempo que demoran en realizar el proyecto (semanas)")
plt.ylabel("Costo de desarrollar el proyecto (millones de pesos)")
plt.legend()
plt.show()

# Respuesta 2
def ecuacion1(t):
  return 12*t + 50 - 155 # Queremos que C1(t) = 155

t1 = np.linspace(0, 200, 1) # Valor inicial para que fsolve arranque

solucion = fsolve(ecuacion1, t1) # Resolver

print(f"Respuesta 2: La cantidad de semanas que duró el trabajo para que el costo fuera de 155 millones de pesos fue de {solucion[0]} semanas")

# Respuesta 3
# Punto de equilibrio
def C1(t):
  return 12*t + 50

def C2(t):
  return 8*t + 100


def inter(t):
  return C1(t) - C2(t)

t = np.linspace(0, 100, 2)
solucion = np.around(fsolve(inter,t),decimals=1)
print(f"Respuesta 3: El punto de equilibrio ocurre a las {solucion[1]} semanas. Por lo tanto, desde {solucion[1]} semanas conviene la empresa 2, ya que los costos desde ese momento son menores que la empresa 1")
```
